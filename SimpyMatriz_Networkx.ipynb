```python
import matplotlib.pyplot as plt

# Definir los estados de la cadena de Markov
estados = ['Soleado', 'Nublado', 'Lluvioso']

# (estado_origen, estado_destino, probabilidad)
trans = [
    ('Soleado', 'Soleado', 0.6),
    ('Soleado', 'Nublado', 0.3),
    ('Soleado', 'Lluvioso', 0.1),
    ('Nublado', 'Soleado', 0.2),
    ('Nublado', 'Nublado', 0.5),
    ('Nublado', 'Lluvioso', 0.3),
    ('Lluvioso', 'Soleado', 0.1),
    ('Lluvioso', 'Nublado', 0.4),
    ('Lluvioso', 'Lluvioso', 0.5)
]

# Crear un grafo dirigido
G_markov = nx.DiGraph()

# Añadir los nodos (estados)
G_markov.add_nodes_from(estados)

# Añadir las aristas (transiciones) con sus probabilidades como 'weight'
for origen, destino, proba in trans:
    G_markov.add_edge(origen, destino, weight=proba)

# Visualizar el grafo
pos = nx.circular_layout(G_markov) # Puedes probar otras distribuciones como nx.spring_layout, nx.shell_layout, etc.

plt.figure(figsize=(8, 6))

nx.draw_networkx_nodes(G_markov, pos, node_size=3000, node_color='lightblue')
nx.draw_networkx_labels(G_markov, pos, font_size=10, font_weight='bold')

# Dibujar las aristas
nx.draw_networkx_edges(G_markov, pos, edgelist=G_markov.edges(),
                       arrowsize=20, width=2, edge_color='gray', alpha=0.7)

# Obtener las etiquetas de las aristas (probabilidades)
edge_labels = nx.get_edge_attributes(G_markov, 'weight')

nx.draw_networkx_edge_labels(G_markov, pos, edge_labels=edge_labels, font_color='red')

plt.title('Grafo de Cadena de Markov con Probabilidades de Transición')
plt.axis('off') # Ocultar los ejes
plt.show()
```

# Árbol de expansión mínima


Sea $N={1,2,...,n} \quad $ el conjunto de nodos de la red
$C_k=\text{conjunto de nodos que se han conectado de forma permanente en la iteración k}$
$\overline{C_k}=\text{Conjunto de nodos que todavía se tienen que conectar en forma permanente}$

Paso 0: Hacer $C_0=\phi$ y $\overline{C_0}=N$

Paso 1: Comenzar con cualquier nodo i en $\overline{C_0}$ e igualar $C_1={i}$ y $\overline{C_1}=N-{i}$. Hacer k=2

Paso k: Seleccionar el nodo $j*\in \overline{C_{k-1}}$ que produzca el arco más cortoa un nodo en $C_{k-1}$. Enlazar a $j*$ en forma permanente con $C_{k-1}$ y sacarlo de $\overline{C_{k-1}}$, esto es:
$$
C_k=C_{k-1}+\{j*\} \qquad \overline{C_{k}}=\overline{C_{k-1}}-\{j*\}
$$
Si $\overline{C_{k-1}}=\phi$ detenerse

Otro caso Hacer $k=k+1$ y repetir el paso

Agregamos librerías

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Creamos el grafo hecho en clase

```python
E = nx.Graph()
E.add_edges_from(
    [
        (1, 2, {"weight": 1}),
        (1, 4, {"weight": 7}),
        (1, 6, {"weight": 9}),
        (1, 3, {"weight": 5}),
        (2, 6, {"weight": 3}),
        (2, 3, {"weight": 6}),
        (2, 4, {"weight": 4}),
        (3, 5, {"weight": 10}),
        (3, 4, {"weight": 5}),
        (4, 5, {"weight": 3}),
        (4, 6, {"weight": 8}),
    ]
)
```

```python
T = nx.minimum_spanning_tree(E)
```

```python
# Visualizamos gráfico de expansión mínima
pos = nx.spring_layout(E)
nx.draw_networkx_nodes(E, pos, node_color="lightblue", node_size=500)
nx.draw_networkx_edges(E, pos, edge_color="grey")
nx.draw_networkx_labels(E, pos, font_size=12, font_family="sans-serif")
nx.draw_networkx_edge_labels(
    E, pos, edge_labels={(u, v): d["weight"] for u, v, d in E.edges(data=True)}
)
nx.draw_networkx_edges(T, pos, edge_color="green", width=2)
plt.axis("off")
plt.show()
```

# Problema de la ruta más corta

## **Algoritmo de Dijkstra**

```python
#librerías
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Creamos el gráfico con los nodos y sus pesos
D = nx.DiGraph()
D.add_nodes_from(["1", "2", "3", "4", "5"])
D.add_edge("1", "2", weight=100)
D.add_edge("1", "3", weight=30)
D.add_edge("2", "3", weight=20)
D.add_edge("4", "2", weight=15)
D.add_edge("3", "4", weight=10)
D.add_edge("3", "5", weight=60)
D.add_edge("4", "5", weight=50)
```

Encontramos la ruta más corta con:

```python
corto = nx.shortest_path(D, "1", "5", weight="weight")
```

Imprimimos longitud y la ruta más corta con lo siguiente:

```python
longitud = nx.shortest_path_length(D, "1", "5", weight='weight')
print(f"Longitud total: {longitud}")
print(f"Ruta más corta: {' → '.join(corto)}")
```

Para insertar el dibujo de nuestra ruta más corta, nos suguiere networkx

```python
# Crea una lista de aristas en la ruta más corta
corto_borde = list(zip(corto, corto[1:]))
# Lista de todos los bordes y asigna colores a los que están en la ruta
color_borde = [
    "red" if edge in corto_borde or tuple(reversed(edge)) in corto_borde else "black"
    for edge in D.edges()
]
```

```python
#Visualización del gráfico
pos = nx.spring_layout(D)
nx.draw_networkx_nodes(D, pos)
nx.draw_networkx_edges(D, pos, edge_color=color_borde)
nx.draw_networkx_labels(D, pos)
nx.draw_networkx_edge_labels(
    D, pos, edge_labels={(u, v): d["weight"] for u, v, d in D.edges(data=True)}
)

plt.show()
```

# Problema del Flujo Máximo

1. Identificar una trayectoria de aumento en la red residual. **Si no existe, el patrón de flujo ya es óptimo**

2. Identificar la capacidad residual de $C^*$ de esta trayectoria de aumento. **Se aumenta en C^* el flujo de esta trayectoria**

3. Se disminuye en $C^*$ la capacidad residual de cada arco en esta trayectoria de aumento. Se aumenta en $C^*$ la capacidad residual en la dirección opuesta. Volver al paso 1.

Creamos el grafo a resolver como en los métodos anteriores

```python
F = nx.Graph()
F.add_edge("O", "A", capacity=5)
F.add_edge("O", "C", capacity=4)
F.add_edge("O", "B", capacity=7)
F.add_edge("A", "B", capacity=1)
F.add_edge("A", "D", capacity=3)
F.add_edge("B", "C", capacity=2)
F.add_edge("B", "E", capacity=5)
F.add_edge("B", "D", capacity=4)
F.add_edge("C", "E", capacity=4)
F.add_edge("D", "E", capacity=1)
F.add_edge("D", "T", capacity=9)
F.add_edge("E", "T", capacity=6)
```

El siguiente regresa el total de la ruta y un diccionario de todos los flujos

```python
maximo, flujo_dict = nx.maximum_flow(F, "O", "T", capacity='capacity')
print("El flujo máximo es: ", maximo)
```

```python
for u in flujo_dict:
  for v in flujo_dict[u]:
    if flujo_dict[u][v] > 0:
      capacidad = F[u][v]['capacity']
      print(f"  {u} → {v}: {flujo_dict[u][v]}/{capacidad}")
```

Se definen las posiciones:

```python
pos = {"O": (0, 1),"A": (1, 2), "B": (1, 1), "C": (1, 0),"D": (2, 2), "E": (2, 0),"T": (3, 1)}
```

```python
# Crear figurplt.figure(figsize=(14, 8))

# dibujamos los nodos
nx.draw_networkx_nodes(F, pos, node_color='lightblue',
                      node_size=1500, alpha=0.7)

# aristas con flujo >0
aristas_con_flujo = []
aristas_sin_flujo = []

for u in flujo_dict:
    for v in flujo_dict[u]:
        if flujo_dict[u][v] > 0:
            aristas_con_flujo.append((u, v))
        else:
            aristas_sin_flujo.append((u, v))

# aristas sin flujo de gris
nx.draw_networkx_edges(F, pos, edgelist=aristas_sin_flujo,
                      edge_color='gray', width=1, alpha=0.3,
                      arrows=True, arrowsize=15, arrowstyle='->')

# aristas con rojo
nx.draw_networkx_edges(F, pos, edgelist=aristas_con_flujo,
                      edge_color='red', width=3, alpha=0.8,
                      arrows=True, arrowsize=20, arrowstyle='->')

# etiquetas en nodos
nx.draw_networkx_labels(F, pos, font_size=12, font_weight='bold')

# otras etiquetas
edge_labels = {}
for u, v in F.edges():
    capacidad = F[u][v]['capacity']
    flujo_actual = flujo_dict[u].get(v, 0)
    if flujo_actual > 0:
        edge_labels[(u, v)] = f"{flujo_actual}/{capacidad}"
    else:
        edge_labels[(u, v)] = f"0/{capacidad}"

nx.draw_networkx_edge_labels(F, pos, edge_labels=edge_labels,
                           font_size=9, font_weight='bold')

plt.title(f"Flujo Máximo: {maximo} - Rutas de Flujo en Rojo", fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```
